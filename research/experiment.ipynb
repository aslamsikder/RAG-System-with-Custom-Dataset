```python
#Data Ingestion
from langchain_community.document_loaders import CSVLoader
```

```python
file_path = "data/questions.csv"
loader = CSVLoader(file_path)
data = loader.load()
```

```python
data
```

```python
len(data)
```

```python
#Chunking
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 500,
    chunk_overlap = 100
)
documents = text_splitter.split_documents(data)
```

```python
documents
```

```python
len(documents)
```

```python
#Vector Embeddings and Vector store
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(documents,OpenAIEmbeddings())
```

```python
# Chroma Vector Database
query = "নোবেল প্রাপ্ত কাব্যগ্রন্থ 'গীতাঞ্জলি' এর ইংরেজি অনুবাদক কে?"
result = db.similarity_search(query)
result[0].page_content
```

```python
db
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-3.5-turbo-0125",
    temperature=0
)
llm
```

```python
## Design ChatPrompt Template
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""
    Answer and explain the following question based only on the provided context.
    Think step by step before providing a detailed answer and explanation.
    You are the most talented question-answer generator and always helpful.

    - If the question is in Bangla, answer in Bangla.
    - If in English, answer in English.
    - Provide the actual answer, e.g., Option 1: ক) or Option 2: খ) or Option 3: গ) or Option 4: ঘ).

    <context>
    {context}
    </context>
    Question: {input}
    """)
```

```python
# Chain Instruction
# Create stuff Document Chain
from langchain.chains.combine_documents import create_stuff_documents_chain

document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
retriever = db.as_retriever()
retriever
```

```python
# Retriever Chain
from langchain.chains import create_retrieval_chain
retriever_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retriever_chain.invoke({"input":"নোবেল প্রাপ্ত কাব্যগ্রন্থ 'গীতাঞ্জলি' এর ইংরেজি অনুবাদক কে?"})
```
